```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("obesityanalysis.csv")
data.info()
data.isnull().sum()
fig = plt.figure(figsize = (10,8))
features=["Gender","Label"]
for i in range(len(features)):
    plt.subplot(2, 1, i+1)
    ax=sns.countplot(x=features[i],data=data, palette="Set2")
plt.show()
```
